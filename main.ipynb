```python
import pandas as pd

df = pd.read_excel(
    "data/240624 REV0 Data voor Hackathon.xlsx",
    sheet_name="DATA",
    skiprows=6,
    skipfooter=1,
)
# make columns more readable
df.columns = [
    "datetime",
    "offshore_wind_electricity",
    "offshore_wind_speed",
    "offshore_wind_mwh",
    "onshore_wind_electricity",
    "onshore_wind_speed",
    "onshore_wind_mwh",
    "solar_electricity",
    "solar_direct_irradiance",
    "solar_diffused_irradiance",
    "temperature",
    "solar_mwh",
    "total_supply_mwh",
    "peak_demand_mwh",
    "load_percentage",
    "total_demand_mwh",
    "surplus_mwh",
]
df["datetime"] = pd.to_datetime(df["datetime"], format="%m-%d-%Y %H:%M:%S")
df.set_index("datetime", inplace=True)
```

```python
# calculate the percentage of weekly offshore wind energy in the total weekly energy supply
df["offshore_wind_percentage"] = df["offshore_wind_mwh"] / df["total_supply_mwh"] * 100

# calculate the percentage of weekly onshore wind energy in the total weekly energy supply
df["onshore_wind_percentage"] = df["onshore_wind_mwh"] / df["total_supply_mwh"] * 100

# calculate the percentage of weekly solar energy in the total weekly energy supply
df["solar_percentage"] = df["solar_mwh"] / df["total_supply_mwh"] * 100
```

```python
# Calculate weekly sums and directly add them to the DataFrame using transform
df["offshore_wind_mwh_weekly"] = df["offshore_wind_mwh"].resample("W").transform("sum")
df["onshore_wind_mwh_weekly"] = df["onshore_wind_mwh"].resample("W").transform("sum")  # Corrected column name
df["solar_mwh_weekly"] = df["solar_mwh"].resample("W").transform("sum")
df["total_demand_mwh_weekly"] = df["total_demand_mwh"].resample("W").transform("sum")
df["total_supply_mwh_weekly"] = df["total_supply_mwh"].resample("W").transform("sum")
```
